```python
#Finacial impact Analysis
#Goal to understand how vulnerabilities translate to financial losses.
#1.what is the Average financial loss per vulnerability type?
display(df.groupby("Security Vulnerability Type")["Financial Loss (in Million $)"].mean().sort_values(ascending=False))
```

social engineering has the highest average of financial loss across all industries,followed by weak passwords,zero day exploit and lastly unpatched software.

```python
# Determine the financial loss according to resolution time in banking industry?

# Convert 'Resolution Time (in Hours)' to numeric, handling potential errors
df['Incident Resolution Time (in Hours)'] = pd.to_numeric(df['Incident Resolution Time (in Hours)'], errors='coerce')

# Drop rows where 'Resolution Time (days)' is NaN after conversion
df_cleaned = df.dropna(subset=['Incident Resolution Time (in Hours)', 'Financial Loss (in Million $)'])

# Filter data for the Banking industry
banking_df_cleaned = df_cleaned[df_cleaned['Target Industry'] == 'Banking']

# Group by resolution time and calculate the average financial loss
financial_loss_by_resolution = banking_df_cleaned.groupby('Incident Resolution Time (in Hours)')['Financial Loss (in Million $)'].mean().reset_index()

# Generate the line plot
plt.figure(figsize=(5, 4))
sns.lineplot(x='Incident Resolution Time (in Hours)', y='Financial Loss (in Million $)', data=financial_loss_by_resolution)
plt.title('Average Financial Loss vs. Resolution Time in Banking Industry')
plt.xlabel('Resolution Time (in Hours)')
plt.ylabel('Average Financial Loss (in Million $)')
plt.grid(True)
plt.show()
```

It is obvious from the results that the minimum incident resolution time is 1 hour and it cost banks on average financial loss of 79 million $ and the  maximum  incident resolution time is 72  hours and on average cost approximately 37 million $ loss.6o hours of incident resolution has the highest average financial loss of 80 million $.However, there is inconsistency of amount lost between 1 to 72 hours taken to resolve incidents.

```python
# determine the number of affected users during phishing,ransomware and SQL attacks in banking industry, from 2020 to 2024

# Filter data for the banking industry and the years 2020-2024
df_banking_2020_2024 = df[(df['Target Industry'] == 'Banking') & (df['Year'].between(2020, 2024))]

# Filter for specific attack types
attack_types = ['Phishing', 'Ransomware','DDoS','Man-in-the-Middle','Malware', 'SQL Injection']
df_filtered_attacks = df_banking_2020_2024[df_banking_2020_2024['Attack Type'].isin(attack_types)]

# Group by 'Attack Type' and sum the 'Number of Affected Users'
affected_users_by_attack = df_filtered_attacks.groupby('Attack Type')['Number of Affected Users'].sum().reset_index()

# Generate the pie chart
plt.figure(figsize=(5, 4)) # Adjust figure size as needed
plt.pie(affected_users_by_attack['Number of Affected Users'], labels=affected_users_by_attack['Attack Type'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Affected Users by Attack Type in Banking Industry (2020-2024)')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

Between the year 2020 to 2024, phishing attack afftected most users in banking sector,this shows that unsatisfied customers or users of fincial services due to phishing attack is more compared tp malware attacks indicating 10.5% of affected users.Ransomware attack caused 18.2% of affected users,followed by man in the middle attack (17.2%) and DDoS affected 16.0% of users in financial sector globally.
